```python
import os
FILE_DIR = os.getcwd()+'/'
```

```python
#Convert fasta to csv

def fasta_to_csv(fasta_file, csv_file):
    sequences = []
    headers = []
    with open(fasta_file, 'r') as file:
        lines = file.readlines()

    current_sequence = ''
    current_header = ''

    for line in lines:
        line = line.strip()
        if line.startswith('>'):
            
            if current_header != '':
                headers.append(current_header)
                sequences.append(current_sequence)
                current_sequence = ''

            current_header = line[1:]
        else:
            current_sequence += line

    headers.append(current_header)
    sequences.append(current_sequence)

    with open(csv_file, 'w') as file:
        file.write('protein_ID,sequence\n')
        for header, sequence in zip(headers, sequences):
            file.write(f'{header},{sequence}\n')

    print(f'save to {csv_file}')

Fasta_file = FILE_DIR +'Your_seq.fasta'
fasta_to_csv(Fasta_file,FILE_DIR +'Your_seq.csv')


```

```python
import pandas as pd

SOL_protTrans = pd.read_csv(FILE_DIR +'protTrans_prediction_result.csv')
SOL_protTrans = SOL_protTrans.rename(columns={'predict_result':'protTrans_pred'})

All_csv = pd.read_csv(FILE_DIR +'Your_seq.csv')

merge = All_csv.merge(SOL_protTrans,on='sequence')
merge.to_csv('Your_seq_sol.csv')
```
